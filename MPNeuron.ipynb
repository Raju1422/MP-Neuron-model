# MP Neuron Model

```python
for b in range(X_binarised_train.shape[1]+1):
    y_pred_train=[]
    accuracy=0
    for x,y in zip(X_binarised_train,y_train):
        y_pred  = (np.sum(x)>=b)
        y_pred_train.append(y_pred)
        accuracy += (y==y_pred)
    print(b,accuracy, accuracy/X_binarised_train.shape[0])
    
        
```


## From the above the value of b is 27

```python
from sklearn.metrics import accuracy_score
```

# MP Neuron Class

```python
class MPNeuron :
    def __init__(self):
        self.b = None
    def model(self,X):
        return (sum(X)>= self.b)
    ## for predict value of output for each row 
    def predict(self,X):
        Y=[]
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)

    def fit(self,X,Y):
        accuracy ={}
        for b in range(X.shape[1]+1):
            self.b =b
            y_pred = self.predict(X)
            accuracy[b]=accuracy_score(y_pred,Y)
        best_b = max(accuracy,key= accuracy.get)
        self.b = best_b
        print("The value of b is :",best_b)
        print("Highest Accuracy is : " , accuracy[best_b])
        
```

```python
mp_neuron = MPNeuron()
mp_neuron.fit(X_binarised_train,y_train)
```

```python
y_test_pred = mp_neuron.predict(X_binarised_test)
accuracy_score(y_test_pred,y_test)
```
